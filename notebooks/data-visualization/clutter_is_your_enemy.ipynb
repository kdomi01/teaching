```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

```

```python
languages = ["Python", "SQL", "Java", "C++", "JavaScript"]
popularity = [40, 15, 11, 19, 28]
pos = np.arange(len(languages))
```

```python
def set_up_plot():
  barlist = plt.bar(pos, popularity)
  plt.xticks(pos, languages)
  plt.ylabel("popularity in %")
  plt.xlabel("languages")
  plt.title("Most popular languages")

  ax = plt.gca()
  ax.tick_params(axis='both', which='both', length=0)

  for spine in plt.gca().spines.values():
    spine.set_visible(False)

  return barlist
```

```python
set_up_plot();
```

```python
def recolor_bars(barlist, highlight_pos = 0):
  for name, bar in zip(languages, barlist):
   bar.set_color("red")
  barlist[highlight_pos].set_color("navy")
```

```python
colorlist = ["blue", "red", "navy", "grey"]
```

```python
barlist = plt.bar(pos, popularity)
recolor_bars(barlist, highlight_pos = 0)
```

```python
def recolor_bars_to(barlist, colorlist):
  for bar, color in zip(barlist, colorlist):
    bar.set_color(color)
```

```python
recolor_bars_to(barlist, colorlist)
```

```python
def write_labels(bars):
  for bar, pop in zip(bars, popularity):
    print(bar)
    print(bar.get_x())
    lx = bar.get_x() + bar.get_width() / 2
    ly = bar.get_y() + bar.get_height() - 2
    plt.text(lx, ly, pop, color = "white", ha = 'center')
```

```python
set_up_plot()
write_labels(barlist)
```
