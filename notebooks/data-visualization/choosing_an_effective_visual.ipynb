```python
men_p = {"Before September" : 9.3,
         "September" : 5.3,
         "October" : 19.5,
         "November" : 43.3,
         "First 2 weeks of December" : 18.4,
         "Last 2 weeks of December" : 4.2}

women_p = {"Before September" : 15.8,
           "September" : 8.2,
           "October" : 21,
           "November" : 39.9,
           "First 2 weeks of December" : 12.7,
           "Last 2 weeks of December" : 2.4}
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = list(men_p.keys())
men_values = list(men_p.values())
women_values = list(women_p.values())

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, men_values, width, label='% of men', color='skyblue')
bars2 = ax.bar(x + width/2, women_values, width, label='% of women', color='red')

ax.set_title('Shoppers begin shopping for holidays')
ax.set_xticks(x, labels)
ax.tick_params(left = False, bottom = False, axis = "x", labelrotation = 30)
ax.legend()

for spine in ['top', 'right', 'left', 'bottom']:
    ax.spines[spine].set_visible(False)

plt.show();
```
